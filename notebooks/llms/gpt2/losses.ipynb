```python
import math

import matplotlib.pyplot as plt
```

# (Binary) Cross entropy

```python
def binary_cross_entropy(y, p):
    return -(y * math.log(p) + (1 - y) * math.log(1 - p))

xs = [(i / 100) + 1e-9 for i in range(100)]
ys_wrong = [binary_cross_entropy(y=0, p=xi) for xi in xs]
ys_correct = [binary_cross_entropy(y=1, p=xi) for xi in xs]

fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].plot(xs, ys_wrong)
axes[0].set_title("wrong")
axes[1].plot(xs, ys_correct)
axes[1].set_title("correct")
plt.suptitle("binary_cross_entropy")
plt.show()
```
